```python
import math
def can_make(units):
    req = [c * units for c in composition]
    extra = [max(math.ceil(r) - s, 0) for r, s in zip(req, stock)]
    return sum(math.ceil(e * c) for e, c in zip(extra, cost)) <= budget

def findMaximumAlloyUnits(composition, stock, cost, budget):
    lo, hi = 0, 10**18
    while lo < hi:
        mid = lo + (hi - lo) // 2
        if can_make(mid):
            lo = mid + 1
        else:
            hi = mid

    for i in range(lo-1, -1, -1):
        if can_make(i): 
            return i
    return 0
```
